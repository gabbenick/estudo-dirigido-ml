```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import os

# Todos os csvs na database
csv_files = [f for f in os.listdir('./database') if f.endswith('.csv')]

dfs = []

# Iteração em todos os csvs e transformando em df
for file in csv_files:
    file_path = os.path.join('./database', file)
    df = pd.read_csv(file_path)
    dfs.append(df)

# Concat todos os dfs
df = pd.concat(dfs, ignore_index=True)

# Remoção manual de atributos irrelevantes para análise do movimento humano
df = df.drop(df.columns[0], axis=1)
df = df.drop(['EMG_right', 'EMG_left'], axis=1)

# Conversão do df em um único csv
df.to_csv('database.csv', index=False)

print(df.info())

print(df.describe())

print(df['activity'].unique())
```
