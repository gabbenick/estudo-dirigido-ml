```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.datasets import load_iris
from matplotlib.ticker import FormatStrFormatter

# load Iris dataset
data = load_iris(as_frame=True)

# transform the dataset to a DataFrame
df = data.frame

# print DataFrame to see the structure
print(df)

# make a subplot to insert 2 plots in one figure
fig, axs = plt.subplots(1, 2, figsize=(12, 6))

# plot boxplot with outliers
axs[0].boxplot(df["sepal width (cm)"])
axs[0].set_title('Largura da sépala com outliers')
axs[0].set_ylabel('Largura (cm)')

# plot boxplot without outliers
axs[1].boxplot(df["sepal width (cm)"], showfliers=False)
axs[1].set_title('Largura da sépala sem outliers')
axs[1].set_ylabel('Largura (cm)')

plt.show()
```

```python
fig, axs = plt.subplots(2, 2, figsize=(8, 8))

# plot histograms for all variables

axs[0, 0].hist(df["sepal length (cm)"], range=(4, 8), bins=8, edgecolor='black')
axs[0, 0].set_title('Tamanho da sépala')
axs[0, 0].set_xlabel('Tamanho (cm)')
axs[0, 0].set_ylabel('Quantidade')

axs[0, 1].hist(df["sepal width (cm)"], bins=12, edgecolor='black')
axs[0, 1].set_title('Largura da sépala')
axs[0, 1].set_xlabel('Largura (cm)')
axs[0, 1].set_ylabel('Quantidade')

axs[1, 0].hist(df["petal length (cm)"], bins=12, edgecolor='black')
axs[1, 0].set_title('Tamanho da pétala')
axs[1, 0].set_xlabel('Tamanho (cm)')
axs[1, 0].set_ylabel('Quantidade')

axs[1, 1].hist(df["petal width (cm)"], range=(0, 3), edgecolor='black')
axs[1, 1].set_title('Largura da pétala')
axs[1, 1].set_xlabel('Largura (cm)')
axs[1, 1].set_ylabel('Quantidade')

plt.tight_layout()
plt.show()
```

```python
fig, axs = plt.subplots(2, 2, figsize=(10, 10))

# Scatter plot width vs length

axs[0, 1].scatter(df["sepal width (cm)"], df["sepal length (cm)"], color='salmon', alpha=0.7)
axs[0, 1].set_title('Tamanho da sépala vs Largura da sépala')
axs[0, 1].set_xlabel('Largura da sépala (cm)')
axs[0, 1].set_ylabel('Tamanho da sépala (cm)')

axs[1, 0].scatter(df["sepal length (cm)"], df["sepal width (cm)"], color='green', alpha=0.7)
axs[1, 0].set_title('Largura da sépala vs Tamanho da sépala')
axs[1, 0].set_xlabel('Tamanho da sépala (cm)')
axs[1, 0].set_ylabel('Largura da sépala (cm)')

plt.tight_layout()
plt.show()
```

```python
# attributes from the dataframe
attributes = ['sepal length (cm)', 'sepal width (cm)', 'petal length (cm)', 'petal width (cm)']

corr = df[attributes].corr()

# heatmap with numbers
plt.figure(figsize=(8, 6))
plt.imshow(corr, cmap='coolwarm', interpolation='nearest')

for i in range(len(attributes)):
    for j in range(len(attributes)):
        plt.text(j, i, f'{corr.iloc[i, j]:.2f}', ha='center', va='center', color='black')

plt.colorbar(label='Correlação')
plt.title('Heatmap do Iris dataset')
plt.xticks(range(len(attributes)), attributes, rotation=45)
plt.yticks(range(len(attributes)), attributes)
plt.show()
```
